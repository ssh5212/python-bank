# matplotlib 사용법

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()

plt.plot([1,2,3],[1,2,9])

plt.show()
```

```python
data_dict = {'data_x': [1, 2, 3, 4, 5], 'data_y': [2, 3, 5, 10, 8]}

plt.plot('data_x', 'data_y', data=data_dict)
plt.show()
```

```python
data_x = [1, 2, 3, 4, 5]
data_y = [2, 3, 5, 10, 8]


plt.figure(figsize=(10, 10))
plt.plot(data_x, data_y)
plt.show()
```
